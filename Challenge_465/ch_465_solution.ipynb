### 1/ Calculate the youngest, oldest, and average age of bakers for each gender.

```python
baker_df = pd.read_csv(r"C:\Users\LeLuu\Documents\Python_Practice\Challenge_465\Data\Bakers.csv")
```

```python
baker_df
```

```python
ans1 = baker_df.groupby('Gender').agg(
    Num_of_Bakers= ("Baker","count"),
    Youngest_Baker=("Age","min"),
    Oldest_Baker=("Age","max"),
    Avg_age=("Age","mean"),
    ).reset_index()
```

### 2/ Count how many episodes were in each season and determine which season had the fewest episodes.

```python
episode_df
```

```python
episode_count = episode_df.groupby('Season').size().reset_index(name='Num_of_Episodes')
```

```python
#find minimum number of episodes
min_episodes = episode_count[episode_count['Num_of_Episodes'] == episode_count['Num_of_Episodes'].min()]
```

```python
min_episodes
```

```python
#use iloc to find minimum number of episodes
min_episodes = episode_count.iloc[episode_count['Num_of_Episodes'].idxmin()]
```

### 3/ Determine the top three baking theme categories across the regular episodes of the 14 seasons. Note that the last episode of each season is a special episode final bake-off.

```python
episode_list = episode_df.groupby('Season')['Episode'].apply(list).reset_index()
```

```python
episode_list
```

```python
# 2 ways to do: 
# 1/ Apply the function to remove the last item in the Episode list for each season
# 2/ Apply filter in the Theme to exclude the "final" theme
episode_list['regular'] = episode_list.apply(lambda x: x['Episode'][:-1], axis=1)
```

```python
#Explode the regular episodes into separate rows
episode_list = episode_list.explode('regular')
```

```python
episode_list
```

```python
merge_df = pd.merge(episode_list, episode_df, left_on=['Season','regular'], right_on=['Season','Episode'], how='left')
```

```python
merge_df.groupby('Theme').size().reset_index(name='Count').sort_values(by='Count', ascending=False).head(3)
```

### 4/ Combine the Outcomes and Seasons datasets to find out the name of the winner who had the most star baker and handshakes combined. HINT: Beware of duplicate values in the Seasons dataset!

```python
outcome_df = pd.read_csv(r"C:\Users\LeLuu\Documents\Python_Practice\Challenge_465\Data\Outcomes.csv")
```

```python
seasons_df = pd.read_csv(r"C:\Users\LeLuu\Documents\Python_Practice\Challenge_465\Data\Seasons.csv")
```

```python
outcome_df
```

```python
seasons_df
```

```python
seasons_df.drop_duplicates(subset=['Season', 'Winner'], inplace=True)
```

```python
seasons_df
```

```python
outcome_df['Season'] = outcome_df['SeasonEpisode'].str.extract('s(\d+)e\d+').astype(int)
```

```python
outcome_df
```

```python
merge_df = pd.merge(outcome_df, seasons_df, left_on=['Season','Baker'], right_on=['Season','Winner'], how='inner')
```

```python
merge_df
```

```python
ans_4 = merge_df.groupby(['Season','Baker']).agg(
    Sum_Star_Baker=('Star Baker', 'sum'),
    Sum_Handshake=('Handshake', 'sum')).reset_index()
```

```python
ans_4
```

```python
ans_4['Star_and_Handshake'] = ans_4['Sum_Star_Baker'] + ans_4['Sum_Handshake']
```
