```python
sales_no_null = sales_figures_df.dropna(subset=['Sales'])
```

```python
sales_no_null.dtypes
```

```python
sales_no_null['year_market'] = sales_no_null['Date'].str[:4]
```

```python
sales_no_null
```

```python
sales_convert_ccy_df = pd.merge(sales_no_null,exchange_rates_df,on="Currency",how="inner")
```

```python
sales_convert_ccy_df
```

```python
sales_convert_ccy_df['sales_convert_EUR'] = sales_convert_ccy_df['Sales']*sales_convert_ccy_df['EUR']
```

```python
sales_df = sales_convert_ccy_df.groupby(by=['Segment',"Region","year_market"])['sales_convert_EUR'].sum()
```

```python
sales_df = sales_df.reset_index()
```

```python
sales_df
```

```python
pivot_df = sales_df.pivot(index=['Region','Segment'],columns='year_market',values='sales_convert_EUR')
```

```python
pivot_df=pivot_df.reset_index()
```

```python
pivot_df
```

```python
pivot_df['YOY'] = pivot_df['2023']-pivot_df['2022']
```

```python
pivot_df
```

```python
pivot_df = pivot_df[pivot_df['YOY']<0]
```

```python
pivot_df
```

```python
pivot_df['count_region'] = pivot_df.groupby(['Segment'])['Region'].transform('count')

```

```python
pivot_df
```

```python
second_ans_df = pivot_df[pivot_df['count_region']==3]['Segment']
```
