```python
import pandas as pd
```

```python
df['id'] = df['Field_1'].str.extract(r'^(\d+)\.')
```

```python
df
```

```python
df['id'] =df['id'].ffill()
```

```python
df
```

```python
df['Stadium'] = df['Field_1'].str.extract(r'\d+\.\s(.*)\,.*')
```

```python
df['City'] = df['Field_1'].str.extract(r'\d+\.\s.*\,(.*)')
```

```python
df['Capacity'] = df['Field_1'].str.extract(r'Capacity:\s(\d+\,\d+)')
```

```python
df['Games'] = df['Field_1'].str.extract(r'Games:\s(.*)')
```

```python
df[['Stadium', 'City', 'Capacity']] = df.groupby('id')[['Stadium','City','Capacity']].ffill()
```

```python
df= df[df['Games'].notna()]
```

```python
df
```

```python
df['Games']= df['Games'].str.split('; ')
```

```python
df = df.explode('Games')
```

```python
df
```

```python
df['Date'] = df['Games'].str.extract(r'.*\((.*)\)')+", 2018"
```

```python
df['Games'] = df['Games'].str.replace(r'\s*\(.*\)', '', regex=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d %B, %Y')
```

```python
df =df.drop(columns=['Field_1', 'id'])
```

```python
df = df[['Date','Stadium', 'City', 'Capacity', 'Games']].sort_values(by='Date').reset_index(drop=True)
```
