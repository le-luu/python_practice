```python
#Convert the Date column to datetime format
user_group_df['Date']=pd.to_datetime(user_group_df['Date'], format='%Y-%m-%d')
```

```python
#Transform the Date column to Month Year format
user_group_df['Year_Text'] = user_group_df['Date'].dt.strftime('%B %Y')
```

```python
#Filter the data for March 2025 and June 2023
user_group_df_filter = user_group_df[(user_group_df['Year_Text']=='March 2025') | (user_group_df['Year_Text']=='June 2023')]
```

```python
#Get the min date for each URL details and with March 2025 and June 2023
user_group_df_filter['min_date'] = user_group_df_filter.groupby(['URL_detail','Year_Text'])['Date'].transform('min')
```

```python
#Filter the data for only the min_date
df = user_group_df_filter[user_group_df_filter['min_date']==user_group_df_filter['Date']]
```

```python
df
```

## Question 1:

Calculate the total number of User Groups, the total number of members, and the percentage increase in membership between June 2023 and March 2025.

```python
#Count number of user groups from counting URL_detail and sum of members for each Year_Text
q1_df = df.groupby('Year_Text').agg({'URL_detail':'count','Members':'sum'}).reset_index()
```

```python
q1_df
```

```python
#Pivot the data for sum of Members
mem_df = pd.pivot_table(q1_df,values=['Members'],columns=['Year_Text'],aggfunc='sum').rename(columns={'June 2023':'Total Members June 2023','March 2025':'Total Members March 2025'}).reset_index(drop=True)
```

```python
mem_df
```

```python
#Pivot the data for counting number of User Groups
usercount_df = pd.pivot_table(q1_df,values='URL_detail',columns='Year_Text',aggfunc='sum').rename(columns={'June 2023':'Total User Groups June 2023','March 2025':'Total User Groups March 2025'}).reset_index(drop=True)
```

```python
usercount_df
```

```python
#Concatenate 2 dataframes together
q1_df = pd.concat([usercount_df,mem_df],axis=1)
```

```python
q1_df
```

```python
#Calculate the increase ratio from total members from June 2023 to March 2025
q1_df['Total Members Increase Ratio'] = (q1_df['Total Members March 2025']/q1_df['Total Members June 2023']).round(2)
```

```python
q1_df
```

## Question 2:

Identify the top five User Groups with the largest increase in members.

```python
#Group by URL_detail, UserGroupName, Region and Year_Text to get sum of Members
q2_df = df.groupby(['URL_detail','UserGroupName','Region','Year_Text']).agg({'Members':'sum'}).reset_index()
```

```python
q2_df
```

```python
#Pivot the table for June 2023 and March 2025 Members to get the increase number later
p2_pivot_df = q2_df.pivot_table(values='Members',index=['URL_detail','UserGroupName','Region'],columns='Year_Text',aggfunc='sum').reset_index()
```

```python
p2_pivot_df
```

```python
#Calculate the increase in members from June 2023 to March 2025
p2_pivot_df['Increase Members'] = (p2_pivot_df['March 2025'] - p2_pivot_df['June 2023'])
```

```python
#Sort the Increase Members column to get top 5 User Groups with largest increase in members
p2_pivot_df.sort_values(by='Increase Members',ascending=False).head(5)
```
