```python
import pandas as pd
import json
```

```python
#Read the excel file
json_data_df = pd.read_excel(r"C:\Users\LeLuu\Documents\Python_Practice\Challenge_383\InputFile.xlsx")
```

```python
json_data_df
```

```python
#Apply the json.loads function to parse JSON for the PAYLOAD column
json_data_df['parsed'] = json_data_df['PAYLOAD'].apply(json.loads)
```

```python
json_data_df
```

```python
#Flatten the parsed JSON data into a DataFrame
df = pd.json_normalize(json_data_df['parsed'],max_level=1)
```

```python
df
```

```python
#Doing the same for payload column to parse JSON and normalize it
df['parsed'] = df['payLoad'].apply(json.loads)
data_df = pd.json_normalize(df['parsed'], max_level=1)
```

```python
data_df
```

```python
#Use the explode method to expand the 'pricing' column into separate rows
data_df_expanded = data_df.explode('pricing').reset_index(drop=True)
```

```python
data_df_expanded
```

```python
#Flatten the pricing column
pricing_df = pd.json_normalize(data_df_expanded['pricing'])
```

```python
pricing_df
```

```python
#Prepare the data for final output
final_df = pd.concat([data_df_expanded.drop(columns='pricing'), pricing_df], axis=1).drop(columns=['conditionTable', 'application','salesOrganization','distributionChannel','salesUnit','priceRequired','pricingItem.rateUnit','pricingItem.conditionPricingUnit','pricingItem.conditionUnit']).rename(columns={'pricingItem.rate': 'rate'})
```

```python
final_df
```

```python
final_df = final_df[['conditionType','conditionRecordNumber','materialNumber','rate','validFromDate','validToDate']]
```

```python
#Convert data types
final_df = final_df.astype({'conditionRecordNumber': 'int', 'rate': 'float'})
```

```python
#Remove leading zeros from materialNumber
final_df['materialNumber'] = final_df['materialNumber'].apply(lambda x: x.lstrip('0') if isinstance(x, str) and x.isdigit() else x)
```

```python
#Format the date columns to YYYY-MM-DD
final_df[['validFromDate','validToDate']] = final_df[['validFromDate','validToDate']].applymap(lambda x: f"{x[:4]}-{x[4:6]}-{x[6:]}" if isinstance(x, str) and len(x) == 8 else x)
```
