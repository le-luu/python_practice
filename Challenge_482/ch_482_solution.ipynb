```python
import pandas as pd
import os
```

```python
# Join all CSV files in the folder path to a list
all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.csv')]
```

```python
# Create an empty list to store DataFrames
df_list = []

# Loop through each file, read it into a DataFrame, and append to the list
for file in all_files:
    df = pd.read_csv(file)
    df['source_file'] = os.path.basename(file)  # Add a column with the source file name
    df_list.append(df)

# Concatenate all DataFrames in the list
merged_df = pd.concat(df_list, ignore_index=True)
```

### 1. By how much has each company’s share price increased or decreased over the full 25-year period?

```python
merged_df.dtypes
```

```python
#Convert the 'Date' column to datetime format
merged_df['Date']= pd.to_datetime(merged_df['Date'], format='%Y-%m-%d')
```

```python
#Extract the company name from the source file name
merged_df['company'] = merged_df['source_file'].str.split('_').str[0]
```

```python
#Find the minDate and maxDate for each company
min_max_date_df = merged_df.groupby('company').agg(
    minDate=('Date','min'),
    maxDate=('Date','max')
).reset_index()
```

```python
min_max_date_df
```

```python
# Inner Join min_max_date_df with merged_df on minDate and company
q1_df = min_max_date_df.merge(merged_df, left_on=['minDate','company'], right_on=['Date','company'], how='inner').rename(columns={'Close':'Min_Data_Close'})
```

```python
q1_df.drop(columns=['Date', 'Open','High','Low','Adj Close','Volume','source_file'], axis=1, inplace=True)
```

```python
q1_df
```

```python
q1_df = q1_df.merge(merged_df, left_on=['maxDate','company'], right_on=['Date','company'], how='inner').rename(columns={'Close':'Max_Data_Close'})
```

```python
q1_df.drop(columns=['Date', 'Open','High','Low','Adj Close','Volume','source_file'], axis=1, inplace=True)
```

```python
q1_df
```

```python
q1_df['Percentage_gain'] = round(((q1_df['Max_Data_Close'] - q1_df['Min_Data_Close']) / q1_df['Min_Data_Close']) * 100,2)
```

```python
q1_df.sort_values(by='Percentage_gain', ascending=False, inplace=True)
```

```python
q1_df
```

### 2. Which year saw the highest cumulative gain for each company?

```python
# Extract the year from the 'Date' column
merged_df['Year'] = merged_df['Date'].dt.year
```

```python
merged_df
```

```python
# Sort the DataFrame by 'Date' by ascending order
merged_df.sort_values(by=['Date'], inplace=True)
```

```python
# Find the difference in 'Close' prices for each company (by 'Date' in ascending order)
merged_df['close_diff'] = merged_df.groupby(['company'])['Close'].diff()
```

```python
merged_df
```

```python
# Drop rows with NaN values in 'close_diff' column
merged_df.dropna(subset=['close_diff'], inplace=True)
```

```python
# Group by 'Year' and 'company' to find the cumulative sum of 'close_diff'
cumm_sum_df = merged_df.groupby(['Year','company'])['close_diff'].sum().reset_index()
```

```python
cumm_sum_df
```

```python
# Rank the cumulative sum of 'close_diff' for each company
cumm_sum_df['rank'] = cumm_sum_df.groupby('company')['close_diff'].rank(ascending=False)
```

```python
cumm_sum_df
```

```python
# Filter the DataFrame to get the top rank for each company
cumm_sum_df[cumm_sum_df['rank'] == 1].sort_values(by='company', ascending=True).reset_index(drop=True)
```

### 3. What is the average gain or loss for each company by year-month?

```python
merged_df
```

```python
merged_df['year_month'] = merged_df['Date'].dt.to_period('M')
```

```python
q3_df = merged_df.groupby(['company', 'year_month'], as_index=False)['close_diff'].mean().sort_values(by=['company', 'year_month'], ascending=[True, True])
```

### 4. Which months show the largest gains and losses for each company?


```python
merged_df
```

```python
merged_df['month_str'] = merged_df['Date'].dt.strftime('%b')
```

```python
gain_loss_by_month_df = merged_df.groupby(['company', 'month_str'], as_index=False).agg(
    total_gain_loss=('close_diff', 'sum'))
```

```python
gain_loss_by_month_df
```

```python
largest_month_df = gain_loss_by_month_df.groupby('company').apply(lambda x: x.nlargest(1, 'total_gain_loss')).reset_index(drop=True)
```

```python
largest_month_df
```

```python
smallest_month_df = gain_loss_by_month_df.groupby('company').apply(lambda x: x.nsmallest(1, 'total_gain_loss')).reset_index(drop=True)
```

```python
smallest_month_df
```

```python
q4_df = pd.merge(largest_month_df, smallest_month_df, on='company').rename(columns={
    'month_str_x': 'HighestAvgMonth',
    'month_str_y': 'LowestAvgMonth'
})
```

```python
q4_df.drop(columns=['total_gain_loss_x', 'total_gain_loss_y'], inplace=True)
```

```python
q4_df
```

### 5. Using the Open, Close, High, and Low values, calculate the volatility for each company.

```python
merged_df
```

```python
columns_to_check = ['Open', 'High', 'Low', 'Close']
merged_df['Highest_Value'] = merged_df[columns_to_check].max(axis=1)
merged_df['Lowest_Value'] = merged_df[columns_to_check].min(axis=1)
merged_df['Variance'] = (merged_df['Highest_Value'] - merged_df['Lowest_Value'])
merged_df['Volatility'] = ((merged_df['Variance'] / merged_df['Lowest_Value'])*100).round(2)
```

```python
q5_df = merged_df.sort_values(by=['company', 'Date'])[['Date', 'Open', 'High', 'Low', 'Close','Adj Close','Volume','source_file','company','close_diff', 'Highest_Value', 'Lowest_Value', 'Variance', 'Volatility']]
```
